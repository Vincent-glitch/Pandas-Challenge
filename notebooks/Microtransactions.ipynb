```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load 
file_to_load = "../data/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

```python
# Building CSV to a Dataframe
purchase_df = purchase_data[[
    "Purchase ID",
    "SN",
    "Age",
    "Gender",
    "Item ID",
    "Item Name",
    "Price"]]
# Basic Calculations
purchase_group = purchase_df["SN"].value_counts()
Player_Count = len(purchase_group)
#Print Calculations
print("The total number of players is " + str(Player_Count)+".")
```

## Purchasing Analysis (Total)

```python
#Number of Unique Items
unique_items = purchase_df["Item Name"].unique()
unique = len(unique_items)
#Average Purchase Price
Total = sum(purchase_df['Price'])
Q = len(purchase_df['Price'])
Average_P = Total / Q
#Total Number of Purchases
Purchase_num = Q
#Total Revenue
Revenue = purchase_df["Price"].sum()
#Print Calculations
print("Number of Unique Items: " +str(unique)+".")
print("Average Purchase Price: $"+"{:.2f}".format(Average_P)+".")
print("Total Number of Purchases: "+str(Q)+".")
print("Total Revenue: $"+"{:.2f}".format(Revenue)+".")
```

## Gender Demographics

```python
# Basic Calculations
percentage = purchase_df['Gender'].value_counts(normalize=True)
value_purchase = purchase_df.loc[:, ["Gender", "SN"]].drop_duplicates()
player_counts = value_purchase["Gender"].value_counts()
Gen_ncount = purchase_df['SN'].nunique() 
Gen_count = purchase_df['SN'].count() 
#uild to DataFrame
gender_df = pd.DataFrame({"Total Players":player_counts, "Percentage":percentage.map("{:,.2%}".format)})
                                                                                #", is for every 3 decimals"
                                                                                #". of what to do after decimal
                                                                                #": means everything to right"
                                                                                #"2 is representative of decimal places
# Display DataFrame                
gender_df
```


## Purchasing Analysis (Gender)

```python
# Basic Calculations
Gender_group = purchase_data.groupby(["Gender"])
Gender_group.count().head()
ave_price = Gender_group["Price"].mean()
purch_count = Gender_group["Gender"].count()
total_value = Gender_group["Price"].sum()
AvePP = round(total_value/player_counts,2)
#uild to DataFrame
gen_purchase_analysis = pd.DataFrame({"Purchase Count":purch_count,
                                     "Average Purchase Price": ave_price,
                                     "Total Purchase Value": total_value,
                                     "Average Purchase Total per Person by Gender":AvePP})
# Minor Data Munging
gen_purchase_analysis["Average Purchase Price"] = gen_purchase_analysis["Average Purchase Price"].map("${:.2f}".format)
gen_purchase_analysis["Total Purchase Value"] = gen_purchase_analysis["Total Purchase Value"].map("${:.2f}".format)
gen_purchase_analysis["Average Purchase Total per Person by Gender"] = gen_purchase_analysis["Average Purchase Total per Person by Gender"].map("${:.2f}".format)
# Display DataFrame
gen_purchase_analysis
```

## Age Demographics

```python
# Basic Calculations
age_df = purchase_df.drop_duplicates(subset="SN")
bins = [0,9.9,14.9,19.9,24.9,29.9,34.9,39.9,100]
age_range = ["<10","10-14","15-19","20-24", "25-29", "30-34","35-39","40+"]
age_group = purchase_df["SN"].value_counts()
purchase_df["Age Range"] = pd.cut(purchase_df["Age"], bins, labels=age_range,include_lowest=True)
age_count = purchase_df.groupby("Age Range").nunique()["SN"]
total = age_count.sum()
percentage_players = age_count / total
# Build to DataFrame
age_analysis = pd.DataFrame({"Total Count":age_count,
                             "Percentage of Players":percentage_players
                            })
# Minor Data Munging
age_analysis["Percentage of Players"] = age_analysis["Percentage of Players"].map("{:,.2%}".format)
# Display DataFrame
age_analysis
```

## Purchasing Analysis (Age)

```python
# Basic Calculations
purchase_count = purchase_df.groupby("Age Range").count()["Purchase ID"]
TPV = purchase_df.groupby("Age Range").sum()["Price"]
Quantity = purchase_df.groupby("Age Range").count()["Price"]
UniquePersons = purchase_df.groupby("Age Range").nunique()["SN"]
Avg_PP = TPV / Quantity
Avg_PPP = TPV / UniquePersons
# Build to DataFrame
age_spend_analysis = pd.DataFrame({"Purchase Count":purchase_count,
                             "Average Purchase Price":Avg_PP,
                              "Total Purchase Value":TPV,
                              "Avg Total Purchase per Person":Avg_PPP
                            })
# Minor Data Munging
age_spend_analysis["Average Purchase Price"] = age_spend_analysis["Average Purchase Price"].map("${:.2f}".format)
age_spend_analysis["Total Purchase Value"] = age_spend_analysis["Total Purchase Value"].map("${:.2f}".format)
age_spend_analysis["Avg Total Purchase per Person"] = age_spend_analysis["Avg Total Purchase per Person"].map("${:.2f}".format)
# Display DataFrame
age_spend_analysis
```
